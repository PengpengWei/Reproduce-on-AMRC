```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
data = scipy.io.loadmat('./results/syn_10000_res.mat')
```

```python
Rht = data['Rht']
RUt = data['RUt']
mistakes_det = data['mistakes_idx_det']
mistakes_rnd = data['mistakes_idx_rnd']
```

```python
Rht_mean = Rht.mean(axis=1)
RUt_mean = RUt.mean(axis=1)
mistakes_det_mean = mistakes_det.mean(axis=1)
mistakes_rnd_mean = mistakes_rnd.mean(axis=1)
```

```python
plt.plot(Rht_mean[:len(Rht_mean)-1], label="Rht")
plt.plot(RUt_mean[:len(RUt_mean)-1], label="RUt")
# plt.plot(mistakes_det_mean, label="Rht_det")
# plt.plot(mistakes_rnd_mean, label="Rht_rnd")
plt.legend()
plt.ylim([0,1])
plt.show()
```
